```python
zeros = torch.zeros(2, 3)
ones = torch.ones(2, 3)

print(zeros)
print(ones)
```

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([10, 20, 30])

print(a + b)
print(b - a)
print(a * b)
print(b / a)
```

```python
A = torch.tensor([
    [1, 2],
    [3, 4]
], dtype=torch.float32)

B = torch.tensor([
    [5, 6],
    [7, 8]
], dtype=torch.float32)

C = torch.matmul(A, B)

print(C)
```
