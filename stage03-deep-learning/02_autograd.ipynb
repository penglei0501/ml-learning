```python
import torch

x = torch.tensor(2.0, requires_grad=True)

print(x)
```

```python
y = x ** 2

print(y)
```

```python
y.backward()

print(x.grad)
```

```python
import torch

x = torch.tensor(3.0, requires_grad=True)

y = 2 * x ** 2 + 3 * x + 1

y.backward()

print("x =", x)
print("y =", y)
print("x 的梯度 =", x.grad)
```

```python
import torch

x = torch.tensor(3.0)
target = torch.tensor(6.0)

w = torch.tensor(1.0, requires_grad=True)

pred = w * x

loss = (pred - target) ** 2

loss.backward()

print("预测值:", pred.item())
print("真实值:", target.item())
print("损失:", loss.item())
print("w 的梯度:", w.grad.item())
```

```python
learning_rate = 0.01

with torch.no_grad():
    w -= learning_rate * w.grad

print("更新后的 w:", w.item())
```

```python
import torch

x = torch.tensor(3.0)
target = torch.tensor(6.0)

w = torch.tensor(1.0, requires_grad=True)

learning_rate = 0.01

for epoch in range(20):
    pred = w * x
    loss = (pred - target) ** 2

    loss.backward()

    with torch.no_grad():
        w -= learning_rate * w.grad

    w.grad.zero_()

    print(f"第 {epoch+1} 轮, w={w.item():.4f}, loss={loss.item():.4f}")
```
