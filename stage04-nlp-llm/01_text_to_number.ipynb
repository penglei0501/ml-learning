```python
sentences = [
    "我 喜欢 机器 学习",
    "我 喜欢 深度 学习",
    "机器 学习 很 有趣"
]

all_tokens = []

for sentence in sentences:
    tokens = sentence.split()
    all_tokens.extend(tokens)

unique_tokens = sorted(set(all_tokens))

vocab = {}

for index, token in enumerate(unique_tokens):
    vocab[token] = index

print(vocab)
```

```python
encoded_sentences = []

for sentence in sentences:
    tokens = sentence.split()
    ids = [vocab[token] for token in tokens]
    encoded_sentences.append(ids)

print(encoded_sentences)
```

```python
vocab = {
    "<UNK>": 0,
    "我": 1,
    "喜欢": 2,
    "机器": 3,
    "学习": 4
}

text = "我 喜欢 人工智能"

tokens = text.split()

ids = [vocab.get(token, vocab["<UNK>"]) for token in tokens]

print(ids)
```

```python
vocab = {
    "<PAD>": 0,
    "<UNK>": 1,
    "我": 2,
    "喜欢": 3,
    "机器": 4,
    "学习": 5,
    "AI": 6
}

sentences = [
    "我 喜欢 机器 学习",
    "我 喜欢 AI"
]

max_len = 4

encoded_sentences = []

for sentence in sentences:
    tokens = sentence.split()
    ids = [vocab.get(token, vocab["<UNK>"]) for token in tokens]

    while len(ids) < max_len:
        ids.append(vocab["<PAD>"])

    encoded_sentences.append(ids)

print(encoded_sentences)
```

```python
import torch

input_ids = torch.tensor(encoded_sentences)

print(input_ids)
print(input_ids.shape)
```
