```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_name = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"

model = SentenceTransformer(model_name)
```

```python
sentences = [
    "我喜欢机器学习。",
    "我正在学习人工智能。",
    "深度学习是机器学习的一个重要方向。",
    "今天广州天气很好。",
    "我晚上想去吃火锅。",
    "AI Agent 可以调用工具完成任务。",
    "RAG 可以让大模型使用外部知识库。"
]

embeddings = model.encode(sentences)

print(type(embeddings))
print(embeddings.shape)
```

```python
def cosine_sim(text1, text2):
    emb1 = model.encode([text1])
    emb2 = model.encode([text2])
    
    score = cosine_similarity(emb1, emb2)[0][0]
    
    return score
```

```python
def normalize_vector(v):
    norm = np.linalg.norm(v)
    return v / norm
```

```python
emb1 = model.encode("AI Agent 可以调用工具完成任务。")
emb2 = model.encode("智能体能够使用外部工具解决问题。")

dot_before = np.dot(emb1, emb2)

emb1_norm = normalize_vector(emb1)
emb2_norm = normalize_vector(emb2)

dot_after = np.dot(emb1_norm, emb2_norm)

cos_score = cosine_similarity([emb1], [emb2])[0][0]

print("归一化前点积：", dot_before)
print("归一化后点积：", dot_after)
print("余弦相似度：", cos_score)
```

```python
document_embeddings = model.encode(documents)

print(document_embeddings.shape)
```

```python
def top_k_search(query, documents, document_embeddings, top_k=3):
    query_embedding = model.encode([query])
    
    scores = cosine_similarity(query_embedding, document_embeddings)[0]
    
    sorted_indices = np.argsort(scores)[::-1]
    
    results = []
    
    for idx in sorted_indices[:top_k]:
        results.append({
            "document": documents[idx],
            "score": scores[idx],
            "index": idx
        })
    
    return results
```

```python
query = "什么是 RAG？"

results = top_k_search(
    query=query,
    documents=documents,
    document_embeddings=document_embeddings,
    top_k=3
)

for r in results:
    print("文档编号：", r["index"])
    print("相似度：", round(r["score"], 4))
    print("文档内容：", r["document"])
    print("-" * 60)
```

```python
def simple_retriever(query, top_k=3):
    results = top_k_search(
        query=query,
        documents=documents,
        document_embeddings=document_embeddings,
        top_k=top_k
    )
    
    print("用户问题：", query)
    print("=" * 80)
    print("检索到的相关内容：")
    print()
    
    for i, r in enumerate(results, start=1):
        print(f"Top {i}")
        print("相似度：", round(r["score"], 4))
        print("内容：", r["document"])
        print("-" * 80)
```

```python
def search_with_threshold(query, documents, document_embeddings, top_k=3, threshold=0.4):
    query_embedding = model.encode([query])
    scores = cosine_similarity(query_embedding, document_embeddings)[0]
    
    sorted_indices = np.argsort(scores)[::-1]
    
    results = []
    
    for idx in sorted_indices:
        if scores[idx] >= threshold:
            results.append({
                "document": documents[idx],
                "score": scores[idx],
                "index": idx
            })
        
        if len(results) >= top_k:
            break
    
    return results
```

```python
query = "RAG 如何检索知识库？"

results = search_with_threshold(
    query=query,
    documents=documents,
    document_embeddings=document_embeddings,
    top_k=3,
    threshold=0.4
)

for r in results:
    print("相似度：", round(r["score"], 4))
    print("内容：", r["document"])
    print("-" * 60)
```

```python
query = "今晚我应该吃什么？"

results = search_with_threshold(
    query=query,
    documents=documents,
    document_embeddings=document_embeddings,
    top_k=3,
    threshold=0.4
)

if len(results) == 0:
    print("没有找到足够相关的内容。")
else:
    for r in results:
        print("相似度：", round(r["score"], 4))
        print("内容：", r["document"])
        print("-" * 60)
```

```python
class SimpleVectorStore:
    def __init__(self, documents, model):
        self.documents = documents
        self.model = model
        self.embeddings = model.encode(documents)
    
    def search(self, query, top_k=3, threshold=None):
        query_embedding = self.model.encode([query])
        scores = cosine_similarity(query_embedding, self.embeddings)[0]
        
        sorted_indices = np.argsort(scores)[::-1]
        
        results = []
        
        for idx in sorted_indices:
            if threshold is not None and scores[idx] < threshold:
                continue
            
            results.append({
                "index": idx,
                "document": self.documents[idx],
                "score": scores[idx]
            })
            
            if len(results) >= top_k:
                break
        
        return results
```

```python
vector_store = SimpleVectorStore(documents, model)
```

```python
query = "什么是生成式语言模型？"

results = vector_store.search(query, top_k=3)

for r in results:
    print("编号：", r["index"])
    print("相似度：", round(r["score"], 4))
    print("内容：", r["document"])
    print("-" * 60)
```

```python
query = "AI Agent 能做什么？"

results = vector_store.search(query, top_k=3)

for r in results:
    print("编号：", r["index"])
    print("相似度：", round(r["score"], 4))
    print("内容：", r["document"])
    print("-" * 60)
```
