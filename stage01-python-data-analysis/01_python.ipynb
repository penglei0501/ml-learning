```python
# 定义计算平均分的函数，并调用函数处理成绩列表
def average_score(scores):
    return sum(scores) / len(scores)

scores = [80, 90, 75, 88, 92]
avg = average_score(scores)

print("平均分:", avg)
```

```python
# 使用字典列表保存并分析多名学生的成绩
students = [
    {"name": "张三", "score": 88},
    {"name": "李四", "score": 95},
    {"name": "王五", "score": 72},
    {"name": "赵六", "score": 60}
]
for x in students:
    print(x["name"], x["score"])
# 计算平均分
average = sum(x["score"] for x in students) / len(students)
print("平均分：", average)

# 找出最高分学生
top_student = max(students, key=lambda x: x["score"])
print("最高分：", top_student["name"], top_student["score"])

# 找出分数大于 80 的学生
print("分数大于 80 的学生：")
for x in students:
    if x["score"] > 80:
        print(x["name"], x["score"])

# 给每个学生增加等级
for x in students:
    if x["score"] >= 90:
        x["grade"] = "优秀"
    elif x["score"] >= 80:
        x["grade"] = "良好"
    elif x["score"] >= 60:
        x["grade"] = "及格"
    else:
        x["grade"] = "不及格"

# 输出最终结果
for x in students:
    print(x["name"], x["score"], x["grade"])
```

```python
# 导入数据分析和绘图所需的工具
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager
```

```python
# 读取后续分析所使用的学生成绩 CSV
df = pd.read_csv("students.csv")
```

```python
# 先查看数据内容，确认 CSV 已正确读取
df.head()
```

```python
# 选出三门课程，并分别计算它们的平均分
subjects = ["math", "english", "computer"]

subject_means = df[subjects].mean()
print("每门课平均分：")
print(subject_means)
```

```python
# 增加总分列
df["total"] = df[subjects].sum(axis=1)

# 找到总分最大值所在的行
top_student = df.loc[df["total"].idxmax()]

print("总分最高的学生：", top_student["name"])
print("总分：", top_student["total"])
```

```python
# 按性别分组，再计算各组每门课程的平均分
gender_means = df.groupby("gender")[subjects].mean()

print("不同性别的平均成绩：")
gender_means
```

```python
# 加载中文字体
font_path = "/home/penglei/.local/share/fonts/NotoSansCJKsc-Regular.otf"
font_manager.fontManager.addfont(font_path)
font_name = font_manager.FontProperties(fname=font_path).get_name()

plt.rcParams["font.sans-serif"] = [font_name]
plt.rcParams["axes.unicode_minus"] = False

# 绘图
subject_means.plot(
    kind="bar",
    color=["skyblue", "orange", "lightgreen"]
)

plt.title("每门课平均分")
plt.xlabel("科目")
plt.ylabel("平均分")
plt.xticks(rotation=0)
plt.ylim(0, 100)
plt.show()
```

```python
# 找出平均分最高的科目，并生成简短分析总结
best_subject = subject_means.idxmax()
best_average = subject_means.max()

print("成绩分析总结：")
print(f"本次一共分析了 {len(df)} 名学生。")
print(f"总分最高的学生是 {top_student['name']}，总分为 {top_student['total']} 分。")
print(f"平均分最高的科目是 {best_subject}，平均分为 {best_average:.2f} 分。")
```
